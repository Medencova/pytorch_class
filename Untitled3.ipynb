```python
import cv2

for n in range(1,41):
    filename = 'py_' + str(n) 
    image = cv2.imread(filename + ".jpg")
    f = open(filename + '.txt')
    i = 0
    for line in f:
        i += 1
        line = line[:-1]
        b = line.split(' ')
    #     print(image.shape)
        width = image.shape[1]
        height = image.shape[0]
    #     print(b)
        x1 = int((float(b[1]) - float(b[3])/2) * width)
        x2 = int((float(b[1]) + float(b[3])/2) * width)
        y1 = int((float(b[2]) - float(b[4])/2) * height)
        y2 = int((float(b[2]) + float(b[4])/2) * height)

    #     print(x1,x2,y1,y2)
    #     cv2.rectangle(image,(x1,y1),(x2,y2),(0,255,0),1)
    #     cv2.imshow('image',image)
    #     k = cv2.waitKey(5)
        crop_img = image[y1:y2, x1:x2]
        print(i)
        cv2.imwrite(filename + '_' + str(i) + ".jpg", crop_img)

```
